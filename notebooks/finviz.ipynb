### Insider Trading Table

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

def get_tables(url):
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    driver.get(url)
    time.sleep(5)

    tables = pd.read_html(driver.page_source)

    driver.quit()

    return tables

url = 'https://finviz.com/insidertrading.ashx'
tables = get_tables(url)

insider_trading_table = tables[5]
```

```python
columns = [
    'ticker',
    'owner',
    'relationship',
    'date',
    'transaction',
    'cost',
    'num_shares',
    'value_dollars',
    'shares_total',
    'SEC_form'
]
# Now insider_trading_table is a DataFrame that you can use for analysis
insider_trading_table = pd.DataFrame(insider_trading_table)
insider_trading_table.columns = insider_trading_table.iloc[0]

# Drop the first row
insider_trading_table = insider_trading_table.drop(insider_trading_table.index[0])

# Reset the index
insider_trading_table.reset_index(drop=True, inplace=True)
```
